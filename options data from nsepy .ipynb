```python
# For data manipulation
import pandas as pd
import numpy as np

# Import get_history function from nsepy module
from nsepy import get_history

# For manipulating date 
from datetime import date

# For plotting
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')

```

```python
# Get options data for NIFTY index
banknifty_options = get_history(symbol='BANKNIFTY',
                            start=date(2021, 11, 26),
                            end=date(2021, 12, 2),
                            index=True,
                            expiry_date=date(2021, 12, 2),
                            option_type='CE',
                            strike_price=36000,
                            )

banknifty_options.head()
```

```python
fig = plt.figure(figsize=(15, 10))

# Plot close prices
ax = fig.add_subplot(211)
ax.plot(banknifty_options['Close'])
ax.set_title('36000 Strike BANKNIFTY Call Close Price', fontsize=14)
ax.set_ylabel('Close Price', fontsize=12)
ax.set_xlabel('Date', fontsize=12)


# Plot the cumulative returns
ax = fig.add_subplot(212)
ax = banknifty_options['Open Interest'].plot(kind='bar', color='c')
ax.set_title('36000 Strike BANKNIFTY Call Open Interest', fontsize=14)
ax.set_ylabel('Open Interest', fontsize=12)
ax.set_xlabel('Date', fontsize=12)

plt.show()
```
